**17.Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip. Hint: Add a column chain to the dataset. A chain can have more than two trips.**

**df1** is a copy of df where a new column location is created, assigned the value of the DOLocationID column (Drop-Off Location ID).
Adds a new column *left_index*, which stores the original row index from df.

**df2** is a copy of df where a new column location is created, assigned the value of the PULocationID column (Pick-Up Location ID).
Adds a new column *right_index*, which stores the original row index from df.

**Sorting df1**: 
The DataFrame is sorted by the following columns, in order of priority:
-tpep_dropoff_datetime (Drop-Off Time)
-VendorID (Vendor Identifier)

**Sorting df2**:
The DataFrame is sorted by the following columns, in order of priority:
-tpep_pickup_datetime (Pick-Up Time)
-VendorID (Vendor Identifier)
-location (Pick-Up Location ID)
-tpep_dropoff_datetime (Drop-Off Time).

The purpose of sorting is to ensure that the rows are ordered logically for subsequent operations, which is a merge.

```python
#  we drop unnecessary columns
columns_to_drop = ['passenger_count', 'trip_distance', 'RatecodeID', 'store_and_fwd_flag',
                   'payment_type', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 
                   'tolls_amount', 'improvement_surcharge', 'total_amount', 'congestion_surcharge']
df = nyc_new.drop(columns=columns_to_drop)

#  we create two separate copies of the dataframe , we create for eache dataframe a new column called location and we create also the columns containing the indexes
df1 = df.assign(location=df['DOLocationID'], left_index=df.index).copy()
df2 = df.assign(location=df['PULocationID'], right_index=df.index).copy()

# we sort each DataFrame and reset indices
df1 = df1.sort_values(by=['tpep_dropoff_datetime', 'VendorID', 'location', 'tpep_pickup_datetime'])
df2 = df2.sort_values(by=['tpep_pickup_datetime', 'VendorID', 'location', 'tpep_dropoff_datetime'])
```

This code performs an **as-of merge** between two DataFrames, df1 and df2, aligning rows based on specific conditions.

**pd.merge_asof** merges rows from df1 and df2 based on the closest matching timestamps.

**left_on & right_on** matches tpep_dropoff_datetime from df1 with tpep_pickup_datetime from df2.

**by** ensures matches occur only when both VendorID and location columns are the same.

**tolerance=pd.Timedelta('2 minutes')** allows matches only if the time difference between the two timestamps is 2 minutes or less.

**direction='forward'** matches only to future timestamps (right_on >= left_on).
The result, merged, contains rows from both DataFrames with the closest valid matches.

```python
merged=pd.merge_asof(
    left=df1,
    right=df2,
    left_on='tpep_dropoff_datetime',
    right_on='tpep_pickup_datetime',
    by=['VendorID', 'location'],
    tolerance=pd.Timedelta('2 minutes'),
direction='forward') 
merged
```

```python
merged=merged.dropna() #we remove the NA from the previous result
merged.head()
```

The purpose is to remove duplicate rows based on the right_index column (which corresponds to the original index from df2).

**subset='right_index'** specifies that duplicates should be identified based on the right_index column.

**keep='first'**  keeps only the first occurrence of each right_index value and drops the subsequent duplicates. This ensures that for each unique right_index, only the first row (the one with the earliest drop-off time) is kept.

```python
merged=merged.drop_duplicates(subset='right_index', keep='first')#we remove the duplicated rows , 
#keeping only the first to retain only those trip
#with the earliest dropoff time before the next departure
merged[['left_index', 'right_index']].sort_values(by='right_index')
```

**df['chain'] = df.index** it initializes a new column chain in the DataFrame df and assigns each row its index as a unique "chain" ID. This means initially, each row is its own chain.

**index_to_chain = df['chain'].to_dict()** converts the chain column to a dictionary, where the keys are the row indices (df.index), and the values are the corresponding chain IDs,which are initially just the indices themselves.

**for _, row in merged.iterrows()** iterates over each row of the merged DataFrame, which contains pairs of rows from df1 and df2 that are being merged.

**left_chain = index_to_chain[row['left_index']] and right_chain = index_to_chain[row['right_index']]** this retrieves the current chain IDs for the rows identified by left_index and right_index from the index_to_chain dictionary.

**new_chain = min(left_chain, right_chain)** for each pair of rows (left and right), the code assigns the smaller of the two chain IDs to both rows, effectively merging the two chains into one.

**index_to_chain[row['left_index']] = new_chain and index_to_chain[row['right_index']] = new_chain** updates the index_to_chain dictionary, assigning the new chain ID to both the left_index and right_index.

**df['chain'] = df.index.map(index_to_chain)** Finally, this updates the chain column in the original DataFrame df with the new chain IDs by mapping the row indices to the corresponding chain IDs from the index_to_chain dictionary.

```python
df['chain'] = df.index  # Assign each row its own chain initially
index_to_chain = df['chain'].to_dict()

# Iteratively update chains
for _, row in merged.iterrows():
    left_chain = index_to_chain[row['left_index']]
    right_chain = index_to_chain[row['right_index']]
    
    # Propagate the smaller chain ID to both left and right
    new_chain = min(left_chain, right_chain)
    index_to_chain[row['left_index']] = new_chain
    index_to_chain[row['right_index']] = new_chain

# Update the DataFrame with the new chain IDs
df['chain'] = df.index.map(index_to_chain)
```

**chain_mapping = {}**  initializes an empty dictionary chain_mapping that will store the final root chain ID for each row index.

**for idx, chain_id in index_to_chain.items():** this loops through each row index (idx) and its corresponding chain ID (chain_id) in the index_to_chain dictionary.

**root_chain = chain_id** initializes root_chain with the current chain_id for the current row.

**while root_chain != index_to_chain[root_chain]:** this is the key part of the process. It checks whether the current chain ID is the root chain for that component. If the chain ID is not the root , it keeps following the chain until it reaches the ultimate root.

**root_chain = index_to_chain[root_chain]** updates root_chain to the chain ID of the current root_chain, effectively following the chain upwards until the root is found.

**chain_mapping[idx] = root_chain** Once the ultimate root is found for the current row, it assigns this root chain ID to the chain_mapping dictionary for that row index (idx).

**df['chain'] = df.index.map(chain_mapping)** this applies the final mapping to the df DataFrame. It maps each row index to its final root chain ID (from chain_mapping), updating the chain column with the ultimate root chain for each row. This ensures that each row is assigned to its true cluster or connected component based on the earlier merging process.

```python
chain_mapping = {}
for idx, chain_id in index_to_chain.items():
    root_chain = chain_id
    # Find the ultimate root for the chain
    while root_chain != index_to_chain[root_chain]:
        root_chain = index_to_chain[root_chain]
    chain_mapping[idx] = root_chain

# Apply the final chain mapping
df['chain'] = df.index.map(chain_mapping)
```

**idxmax()** returns the index (the chain ID) of the maximum value. In other words, it returns the most frequent chain ID, which represents the "largest" or most common group of rows in the chain column.
